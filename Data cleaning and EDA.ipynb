### Data Munging using Pandas library

***Data munging is an exercise of cleaning up the messy data set by replacing irrelevant data with meaningful values***

***Exploration of SAT_Result data set resulted into conclusion that columns Applicants,CR_Average_Score,Math_Average_Score and Writing_Average_Score contains non-numeric value 's', which are making dataset messy.***

***In order to clean the SAT_Result dataset, 's' needs to be replaced by valid values which will keep the data consistent and other column values should remain unaffected***

### How to fill Missing / Inconsistent data?

***There are numerous ways to fill the inconsistent values:***

***1. Replace inconsistent values with mean of particular column using dataframe.mean() function***

***2. Replace inconsistent values with NaN values using fillna() function***

***3. Replace inconsistent values with 0***

***4. Removing rows containing inconsistent values***

***5. Removing duplicates if any (in this case not)***

### Data Cleaning method used:

**From the above stated possible solutions the one which we have selected here to clean the data is of replacing all inconsistent values with mean values of respective columns which is the simplest method to keep remaining dataset values unaffected**

```python
SAT_Result.replace('s',np.NaN) # replace 's' values of all columns with NaN value using replace function
SAT_Result["Applicants"]=pd.to_numeric(SAT_Result["Applicants"], errors="coerce") # converting datatype of Applicant column to Numeric
SAT_Result["CR_Average_Score"]=pd.to_numeric(SAT_Result["CR_Average_Score"], errors="coerce") # converting datatype of CR_Average_Score column to Numeric
SAT_Result["Math_Average_Score"]=pd.to_numeric(SAT_Result["Math_Average_Score"], errors="coerce") # converting datatype of Math_Average_Score column to Numeric
SAT_Result["Writing_Average_Score"]=pd.to_numeric(SAT_Result["Writing_Average_Score"], errors="coerce") # converting datatype of Writing_Average_Score column to Numeric

print("Applicants Mean:",int(SAT_Result["Applicants"].mean())) #calculating and  Displaying mean value of Applicant's value 
print("Critical Reading Average Score mean",SAT_Result["CR_Average_Score"].mean()) #calculating and Displaying mean value of CR_Average_Score's values 
print("Math Average Score:",SAT_Result["Math_Average_Score"].mean()) #calculating and Displaying mean value of Math_Average_Score's value
print("Writing Average Score:",SAT_Result["Writing_Average_Score"].mean()) #calculating and  Displaying mean value of Writing_Average_Score's value 
```

*** Explanation : In data set SAT_Result, 4 columns Applicants, CR_Average_Score, Math_Average_Score and Writing_Average_Score contain inconsistent value as 's' which needs to be replaced with meaningful data to keep other values unaffected.***

***In order to replace 's' values, replace funcion is used to replace 's' with NaN values via np.NaN function of numpy***

***Furthermore, all the above mentioned columns have been type-casted to numeric values using pd.to_numeric function***

***Mean values of all columns have been calculated using dataframe.mean() function which needs to be use further to replace 's' values with respective mean values***

```python
SAT_Result["Applicants"].fillna(int(SAT_Result["Applicants"].mean()), inplace=True) # replace NaN values of Applicants column with mean value of same column
SAT_Result["CR_Average_Score"].fillna(int(SAT_Result["CR_Average_Score"].mean()), inplace=True)# replace NaN values of Applicants column with mean value of same columnce=True) # replace NaN values of CR_Average_Score column with mean value of same column
SAT_Result["Math_Average_Score"].fillna(int(SAT_Result["Math_Average_Score"].mean()), inplace=True)# replace NaN values of Math_Average_Score column with mean value of same column
SAT_Result["Writing_Average_Score"].fillna(int(SAT_Result["Writing_Average_Score"].mean()), inplace=True)# replace NaN values of Writing_Average_Score column with mean value of same column

SAT_Result # displaying entire dataset
```

*** Explanation: In the above code block all the NaN values have been replaced with mean values of respective columns using fillna() function which replaces NaN values with mentioned values(here mentioned values are mean of respective columns)***

***Now the entire data set contains consistent values***

### Melting data of SAT_Result

***Melting function can be used to unpivot (i.e. it converts columns to rows) the Dataframe 'SAT_Result' from  wide format to long format by associating each measured variables to its subsequent values in an individual row***

```python
tidy = pd.melt(SAT_Result.reset_index(),
               id_vars=['Serial Number', 'SCHOOL'], value_vars=['CR_Average_Score','Math_Average_Score','Writing_Average_Score'],
               value_name='Scores',var_name='Average Score areas')
tidy.head(600)
```

***Explanation: 
Here, Serial Number and SCHOOL are two identifier variables &'CR_Average_Score','Math_Average_Score','Writing_Average_Score' are measured variables which are undergoing process of melting where the function melts down the columns CR_Average_Score,Math_Average_Score,Writing_Average_Score into one column with multiple rows. Identifier variables remains unaffected with repetition of same values***

***value_name='Scores' is used to assign resulting column name as 'Scores' and column name of measured variables has set to Average Score areas by assigning var_name as Average Score areas***

### Creating variety of plots for exploring and understanding of data in detail

### Plot 1: Factor Plot (Applicants / School)

***FactorPlot : Factorplot represents the categorical plot on facetfrid***

***Syntax to use FactorPlot:***

***seaborn.factorplot( x=column to be displayed on x axis,y=column to be displayed on y axis, hue=name of variable in data, data=Dataframe, row=categorical variable, col=categorical variable,kind=kind of plot to draw, size= Height of each facet, aspect=width of each facet, orient=orientation of plot(horizontal or vertical), color=color for all elements, palette=colors to use for different levels of hue )***

***Following is an example of FactorPlot using SAT_Result data***

```python
sns.set(style="white") # Setting the aesthetic style of the plots

g = sns.factorplot(x="Applicants", y="SCHOOL",aspect=2.5, size=6,kind='bar', 
                   data=SAT_Result.query('300<Applicants'), palette='colorblind') # ploting graph of Applicants versus SCHOOL
```

***Explanation: Above graph shows the data of number of applicants versus School name for those schools which has number of applicants greater than 300***

***x="Applicants", y="SCHOOL" are declared to plot a graph of applicants and school.***

***data=SAT_Result.query('300< Applicants'): Using dataframe SAT_Result to plot a graph and .query is used to define a criteria for number of applicants***

** Conclusion: From the above graph, We can conclude that 'Brooklyn Technical High School' has the highest number of applicants(approximately 1300)**

### Plot 2: Bar Plot (School / CR_Average_Score)

***Barplot : This plot shows point estimation and confidence intervals as rectangular bars***

***Syntax for Barplot:***

***seaborn.barplot(x=column to be displayed on x axis,y=column to be displayed on y axis, hue=name of variable in data, data=Dataframe, order=Order to plot the categorical levels, units=Identifier of sampling units, orient=Orientation of the plot (vertical or horizontal), color=Color for all of the elements,palette=Colors to use for the different levels of the hue variable, saturation=Proportion of the original saturation to draw colors, errwidth=Thickness of error bar lines,
capsize=None, dodge=whether elements should be shifted along the categorical axis(true or false))***

***Following is an example of BarPlot using SAT_Result data***

```python
fig, ax = plt.subplots(figsize=(23, 7)) # defining the size of figure
sns.barplot(x='SCHOOL', y='CR_Average_Score', data=SAT_Result.query('623 < CR_Average_Score'),
            color='k', ax=ax) # creating barplot of School versus CR_Average_Score
sns.despine() # Removing the top and right spines from plot
```

***Explanation: Above graph shows the data of School versus CR_Average_Score for the schools which has CR_Average_Score is greater than 623***

***x='SCHOOL'and y='CR_Average_Score' are declared to plot a graph of School and CR_Average_Score***

***data=SAT_Result.query('623 < CR_Average_Score'): Using dataframe SAT_Result to plot a graph and .query is used to define a criteria for CR_Average_Score***

***Conclusion: From the above graph it can be concluded that particiants from 'Stuyvesant High School' has highest CR_Average_Score***
