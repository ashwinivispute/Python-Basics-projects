## Question 1

** What could be the future values of CR Average scores based on the CR average scores from SAT Examination 2010?**

### CHI-SQUARE CALCULATION

***Calculating Chi-Square value for CR_Average_Score_2010 and predicted CR_Average_Score value of 2012***

```python
observed = SAT_Result_2010['CR_Average_Score_2010'].mean()
expected = SAT_Result['CR_Average_Score'].mean()
print('Observed value: ', observed)
print('Expected value: ',expected)
chi_squared_stat = (((observed-expected)**2)/expected)
print('Chi-Square value: ',chi_squared_stat)
```

```python
import scipy
from scipy.stats import chisquare
chisquare(SAT_Result_2010['CR_Average_Score_2010'].mean(),f_exp=SAT_Result['CR_Average_Score'].mean())
```

**Interpretation: Here we have calculated the value of Chi square for CR average scores of year 2010 which is considered to be an observed value and future value of CR average is considered to be an expected one.
From the above calculation we got the Chi-Square value:  0.029868605656417 which is too less to be considered and hence it can be concluded that CR average marks of 2012 could be approximately similar to the marks of 2010 and hence there is not a significant difference in the CR average marks of both years**

### LINEAR REGRESSION CALCULATION

***Calculating Coefficients,p-value, R-squared value***

```python
import statsmodels.formula.api as smf
lm= smf.ols(formula='Applicants_2010 ~ CR_Average_Score_2010' , data=SAT_Result_2010).fit()
lm.params # print coefficients
```

***Calculating prediction of CR average score based on the exisiting value from SAT Result 2010***

```python
lm.predict(SAT_Result_2010.CR_Average_Score_2010)
```

**Interpretations: Here we are predicting the future values of CR average score**

***Calculating the p-value of CR average score***

```python
lm.pvalues
```

***Calculating R-Squared value for CR average score***

```python
lm.rsquared
```

***t-test for CR average score***

```python
import scipy
from scipy import stats
stats.ttest_rel(SAT_Result_2010['CR_Average_Score_2010'].head(),SAT_Result['CR_Average_Score'].head())
```

**Interpretation: From the above t -test calculation , we got a p-value as 0.30389138776396163 which is greater than 0.05. In other words, p> 0.05 then SAT result data sets for year 2010 and 2012 are related to each other and in this case null hypothesis cannot be rejected. Hence there is a difference between the means of CR values and we can not conclude here that significant difference exists.**
